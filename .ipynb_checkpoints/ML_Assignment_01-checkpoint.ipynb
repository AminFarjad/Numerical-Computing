## Task2

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

dataset[numerical_features].head()
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))

continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()

# most cars are manufactured in 2010 decade
# as cars are used hence the mileage is large
```

```python
## Lets Find the realtionship between discrete features and price

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['price'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('price')
    plt.title(feature)
    plt.show()

# Some results from the data automatic_transmission cars are more priced
# Damaged cars are low priced
# Car which are not second hand had more value
# Personal use didn't had any effect on prices
# Car with turbos were more priced
# alloy_wheels car were also more priced
# one unique thing was car which required a key to start were more priced then the cars which didn't required it
# It was observed from the data the more accessories your car had the more it was priced
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
# There are a lot of outliers in price table and the remaining continous features had zero in them so ignored them
# but if plotted still engine_size had a very few outliers
```

```python
categorical_features=[feature for feature in dataset.columns if data[feature].dtypes=='O']
categorical_features
```

```python
dataset[categorical_features].head()
```

```python
for feature in categorical_features:
    data=dataset.copy()
    if len(dataset[feature].unique()) > 100:
        continue
    data.groupby(feature)['price'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('price')
    plt.title(feature)
    plt.show()

# from below charts we can clearly see porsche brand was the more expensive brand and suzuki was the most low priced
# Electric fuels cars were the most priced cars and four wheelers were the most priced
# rest categorical features were ignored as they had a lot of categories
```

```python
li=[]
for feature in continuous_feature:
    li.append(feature)
    for feature1 in continuous_feature:
        if feature == feature1 or feature1 in li:
            continue
        else:
            data=dataset.copy()
            plt.scatter(data[feature1],data[feature])
            plt.xlabel(feature1)
            plt.ylabel(feature)
            plt.title(feature1+'-'+feature)
            plt.show()
# From the below plots we can deduce that cars which were manufactured in the 21st century had a lot of mileage 
# hence the cars before 21st century were abandoned or not used frequently as they had very low mileage
# Cars manufactured in the 20th century had cars with large engine_size
```

```python
missing_data = dataset.isnull().sum()
missing_data
```

```python
dataset['engine_size_pres'] = dataset['engine_size'].apply(lambda x: 1 if pd.notna(x) else 0)
plt.figure(figsize=(10, 6))
mean_prices = dataset.groupby('engine_size_pres')['price'].median().reset_index()
sns.barplot(x='engine_size_pres', y='price', data=mean_prices, palette='Set3')
plt.title('Effect of Engine_size_ Presence on Price')
plt.xlabel('Engine_size Presence (0: Missing, 1: Present)')
plt.ylabel('Price')
plt.xticks([0, 1], ['Missing', 'Present'])
plt.show()
dataset.drop('engine_size_pres', axis=1, inplace=True)
# Effect of Engine_size_ can be seen on the price columns so if someone want to have their car price more valued
# they should not show the engine size
```

```python
dataset['first_owner_pres'] = dataset['first_owner'].apply(lambda x: 1 if pd.notna(x) else 0)
plt.figure(figsize=(10, 6))
mean_prices = dataset.groupby('first_owner_pres')['price'].median().reset_index()
sns.barplot(x='first_owner_pres', y='price', data=mean_prices, palette='Set3')
plt.title('Effect of first_owner_ Presence on Price')
plt.xlabel('first_owner Presence (0: Missing, 1: Present)')
plt.ylabel('Price')
plt.xticks([0, 1], ['Missing', 'Present'])
plt.show()
dataset.drop('first_owner_pres', axis=1, inplace=True)
# Effect of first_owner presence doesn't effect the price
```

```python
dataset['color_presence'] = dataset['exterior_color'].apply(lambda x: 1 if pd.notna(x) else 0)
plt.figure(figsize=(10, 6))
mean_prices = dataset.groupby('color_presence')['price'].median().reset_index()
sns.barplot(x='color_presence', y='price', data=mean_prices, palette='Set3')
plt.title('Effect of Exterior Color Presence on Price')
plt.xlabel('Color Presence (0: Missing, 1: Present)')
plt.ylabel('Price')
plt.xticks([0, 1], ['Missing', 'Present'])
plt.show()
# Drop the 'color_presence' column 
dataset.drop('color_presence', axis=1, inplace=True)
# Effect of exterior color can be seen on the price columns but we can't do any transformation on this column
```

```python
dataset['color_presence'] = dataset['interior_color'].apply(lambda x: 1 if pd.notna(x) else 0)
plt.figure(figsize=(10, 6))
mean_prices = dataset.groupby('color_presence')['price'].median().reset_index()
sns.barplot(x='color_presence', y='price', data=mean_prices, palette='Set3')
plt.title('Effect of interior_color Presence on Price')
plt.xlabel('Color Presence (0: Missing, 1: Present)')
plt.ylabel('Price')
plt.xticks([0, 1], ['Missing', 'Present'])
plt.show()

# Drop the 'color_presence' column
dataset.drop('color_presence', axis=1, inplace=True)
#AS we can see it doesn not effect the prices we can  ignore the interior_color 
```

## Task5

```python
features = [f for f in dataset.columns]
numericFeatures = [f for f in features if dataset[f].dtype == 'int' or dataset[f].dtype== 'float']

for feature in numericFeatures:
    print('Feature', feature, 'Skewness:', skew(dataset[feature].dropna()),'Kurtosis:', kurtosis(dataset[feature].dropna()), '\n' )

# Interpretation:

# Skewness:
# If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.
# Skewness > 0: Right-skewed (positively skewed)
# Skewness < 0: Left-skewed (negatively skewed)
# Kurtosis:
# Kurtosis measures the tails and sharpness of the distribution.
# Kurtosis = 3: Normal distribution
# Kurtosis > 3: Leptokurtic (heavy-tailed, sharp peak)
# Kurtosis < 3: Platykurtic (light-tailed, flat peak)

# price is right skewed hence it needs normalization 
# normalization is required on these mileage, engine_size, automatic_transmission 
# else the remaining columns are distributed normally 
```
